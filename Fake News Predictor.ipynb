```python
import numpy as np
import pandas as pd
```

```python
true = pd.read_csv('True.csv')
```

```python
fake = pd.read_csv('Fake.csv')
```

```python
fake.info()
```

```python
true['label']=1
fake['label']=0
```

```python
true.head()
```

```python
fake.head()
```

```python
news = pd.concat([fake,true],axis=0)

```

```python
news.head()
```

```python
news.tail()
```

```python
news = news.drop(['title','subject','date'],axis = 1)
```

```python
news.head()
```

```python
news.isnull().sum()
```

```python
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.drop(['index'],axis=1,inplace=True)
```

```python
news
```

```python
import re
```

```python
def wordopt(text):
    #convert into lowercase
    text = text.lower()
    # remove urls
    text = re.sub(r'https?://\S+|www\.\S+','',text)
    # remove html tags
    text = re.sub(r'<.*?>','',text)
    # remove punctuations
    text = re.sub(r'[^\w\s]','',text)
    # remove digits
    text = re.sub(r'\d','',text)
    # remove newline characters
    text = re.sub(r'\n','',text)
    
    return text

    
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_train
```

```python
xv_test = vectorization.transform(x_test)
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
DTC.fit(xv_train,y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test, y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(xv_train,y_train)
```

```python
pred_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
print(classification_report(y_test,pred_rfc))
```

```python
from sklearn.model_selection import cross_val_score

# Logistic Regression Cross-Validation
lr_cv_scores = cross_val_score(LR, xv_train, y_train, cv=5)
print("Logistic Regression Cross-Validation Scores: ", lr_cv_scores)
print("Logistic Regression Average Score: ", lr_cv_scores.mean())

# Decision Tree Classifier Cross-Validation
dtc_cv_scores = cross_val_score(DTC, xv_train, y_train, cv=5)
print("Decision Tree Cross-Validation Scores: ", dtc_cv_scores)
print("Decision Tree Average Score: ", dtc_cv_scores.mean())

# Random Forest Classifier Cross-Validation
rfc_cv_scores = cross_val_score(rfc, xv_train, y_train, cv=5)
print("Random Forest Cross-Validation Scores: ", rfc_cv_scores)
print("Random Forest Average Score: ", rfc_cv_scores.mean())
```

```python
def output_label(n):
    if n==1:
        print("News is Genuine")
    elif n==0:
        print("Fake News")
```

```python
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = LR.predict(new_xv_test)
    pred_dtc = DTC.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)

    return "\n\nLR Prediction: {} \nDTC Prediction: {} \nrfc Prediction: {}".format(output_label(pred_lr[0]), 
                                                                                   output_label(pred_dtc[0]),
                                                                                   output_label(pred_rfc[0]))
    
    
```

```python
news_article = str(input())
```

```python
manual_testing(news_article)
```
